```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/sample_data/Teen_Mental_Health_Dataset.csv')
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (10, 5)

print("✅ Libraries loaded")
```

```python
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# CELL 4 — Basic info
print("Column names and types:")
print(df.dtypes)
print("\nAny missing values?")
print(df.isnull().sum())
print("\nBasic statistics:")
df.describe()
```

```python
# CELL 5 — Q1: Which platform do teens use most?
platform_counts = df['platform_usage'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=platform_counts.index, y=platform_counts.values, palette="Blues_d")
plt.title("Most Used Social Media Platforms Among Teenagers", fontsize=14)
plt.xlabel("Platform")
plt.ylabel("Number of Users")
plt.tight_layout()
plt.show()

print(platform_counts)
```

```python
# CELL 6 — Q2: Does more social media = less sleep?
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='daily_social_media_hours', y='sleep_hours',
                hue='gender', alpha=0.6)
plt.title("Social Media Hours vs Sleep Hours", fontsize=14)
plt.xlabel("Daily Social Media Hours")
plt.ylabel("Sleep Hours")
plt.tight_layout()
plt.show()

# Correlation number
corr = df['daily_social_media_hours'].corr(df['sleep_hours'])
print(f"Correlation between social media hours and sleep: {corr:.2f}")
print("(Negative number = more social media, less sleep)")
```

```python
# CELL 7 — Q3: Does social media affect academic performance?
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='daily_social_media_hours', y='academic_performance',
                hue='platform_usage', alpha=0.6)
plt.title("Social Media Hours vs Academic Performance", fontsize=14)
plt.xlabel("Daily Social Media Hours")
plt.ylabel("Academic Performance Score")
plt.tight_layout()
plt.show()
```

```python
# CELL 8 — Q4: Stress and anxiety by platform
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=df, x='platform_usage', y='stress_level', ax=axes[0], palette="Set2")
axes[0].set_title("Stress Level by Platform")
axes[0].set_xlabel("Platform")
axes[0].tick_params(axis='x', rotation=45)

sns.boxplot(data=df, x='platform_usage', y='anxiety_level', ax=axes[1], palette="Set3")
axes[1].set_title("Anxiety Level by Platform")
axes[1].set_xlabel("Platform")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# CELL 9 — Q5: Addiction level by gender and age group
df['age_group'] = pd.cut(df['age'], bins=[12, 14, 16, 18, 20],
                          labels=['13-14', '15-16', '17-18', '19-20'])

plt.figure(figsize=(9, 5))
sns.barplot(data=df, x='age_group', y='addiction_level', hue='gender', palette="coolwarm")
plt.title("Social Media Addiction Level by Age Group and Gender", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Average Addiction Level")
plt.tight_layout()
plt.show()
```

```python
# CELL 10 — Q6: Full correlation heatmap
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            linewidths=0.5, square=True)
plt.title("Correlation Between All Numeric Variables", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# CELL 11 — Key Findings Summary (always end with this)
print("=" * 55)
print("       KEY FINDINGS — Teen Social Media & Mental Health")
print("=" * 55)

print(f"\n📱 Most used platform: {df['platform_usage'].mode()[0]}")
print(f"⏱️  Average daily social media hours: {df['daily_social_media_hours'].mean():.1f} hrs")
print(f"😴 Average sleep hours: {df['sleep_hours'].mean():.1f} hrs")
print(f"📚 Average academic performance: {df['academic_performance'].mean():.2f}")
print(f"😰 Average stress level: {df['stress_level'].mean():.1f} / 10")
print(f"🔄 Social media vs sleep correlation: {df['daily_social_media_hours'].corr(df['sleep_hours']):.2f}")
print(f"📉 Social media vs academics correlation: {df['daily_social_media_hours'].corr(df['academic_performance']):.2f}")
print(f"⚠️  Depression cases in dataset: {df['depression_label'].sum()} out of {len(df)}")
```

```python
The data does not paint social media as a simple villain. The correlations are weak — social media hours alone don't directly destroy sleep or grades in a measurable linear way. What the data actually shows is a lifestyle pattern: teenagers who are highly online tend to have disrupted sleep, moderate stress, adequate-but-not-great academic performance, and elevated anxiety. No single variable is catastrophic. But together, they describe a generation operating below their potential — not because of one bad habit, but because of the cumulative weight of always being connected.
The most important psychological insight here is this: the damage from social media is rarely dramatic. It's the slow erosion of attention, sleep quality, in-person connection, and self-perception — happening so gradually that it feels normal.
```
